#### matplotlib

matplotlib provides a wide variety of ploting functionality in an easy to use interface

##### 2D scatter plots

```python
# make a simple line plot
import numpy as np
import matplotlib.pyplot as plt # Assigns the short name plt to the import matplotlib.pyplot
data = np.arange(16)
# if only a 1D array is provided.  The values will be the y-axis and the x-axis will be
# the array index of each value.
plt.plot(data) # make the plot object
plt.xlabel('array index')
plt.show() # display the plot
plt.plot([1, 2, 3, 4], 'r--', [1, 4, 9, 16], 'bs', [10, 24, 54, 12], 'g^') # multiline plot
plt.show()
plt.plot()
plt.show()
```

##### Figures and subplots

```python
import matplotlib.pyplot as plt

# subplots can be done with or without figures
# The subplot commmand takes a 3 digit number as its argument
# The first digit is the number of columns
# The second digit is the number of rows
# The third digit is the subfigure order. (subfigures do not need to be 
# specified in order in the source code.)

#plt.figure()
plt.subplot(234)
plt.plot([10, 14, 15, 18, 21], 'r--')
plt.subplot(232)
plt.plot([1,2,3,4], 'bs')
plt.subplot(233)
plt.plot([4,3,2,1])
plt.show()
```

#### Seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.arange(16)

# after importing seaborn matplotlib commands will produce output with the new defaults.
plt.plot(data)
plt.show()
```

#### 3D plots

##### 3D scatter plot

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

myFig = plt.figure()

# Add the projection type of 3d to the subfigure command
axis = myFig.add_subplot(111, projection='3d')

x = np.arange(10)
y = np.random.randn(1,10)
z = np.random.rand(1,10)

axis.scatter(x, y, z, c='r', marker='o')

# To add another series to the plot call axis.scatter again.
axis.scatter(x+1, y+1, z+1, c='b', marker='x')

plt.show()
```

Serveral other 3D plot types are available.  See https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html for details. 

```python
import numpy as np
import matplotlib.pyplot as plt

testData = np.random.randn(10000)

plt.hist(testData, 50) # optional second argument is the number of bins
plt.show()
```
