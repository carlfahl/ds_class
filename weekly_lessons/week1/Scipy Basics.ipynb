#### Proability Distributions

````python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

ar1 = np.linspace(0,20,500)

# SciPy provides a number of contineous variable distributions
# stats.norm is the standard normal distribution
# stats.

myCDF = stats.norm.cdf(ar1)
myPDF = stats.norm.pdf(ar1)
# Shift the distribution center to 10
newPDF = stats.norm.pdf(ar1, loc=10)

# The stats method gives moments of the distribution
# 

# Use the rvs method to random sample from a distribution

np.random.seed(4576543) # rvs uses the numpy random seed
# sample 200 points from a normal distribution
# centered at 10.
randomSample = stats.norm.rvs(loc=10, size=200) 

plt.subplot(231)
plt.plot(ar1, myCDF)
plt.subplot(232)
plt.plot(ar1, myPDF)
plt.subplot(233)
plt.plot(ar1, newPDF)
plt.subplot(234)
plt.plot(randomSample, 'bs')
plt.subplot(235)
plt.hist(randomSample, bins=20)
plt.show()
````

#### Statisticial Tests

##### T test of one sample

````python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

# If we want to test if a sample is representative of
# a given probability distribution we can perform a t-test 

sample = stats.norm.rvs(loc=10, size=200)

trueMean = stats.norm.mean(loc=10)

stats.ttest_1samp(sample, trueMean)
````

A high P value indicates that our samples are from the same distribution.  

##### T test comparing two samples

````python
from scipy import stats
import numpy as np

sample1 = stats.norm.rvs(loc=10, size=100)
sample2 = stats.norm.rvs(loc=10, size=100)
sample3 = stats.norm.rvs(loc=11, size=100)

print(stats.ttest_ind(sample1, sample2))
print(stats.ttest_ind(sample1, sample3))

stat, pval = stats.ttest_ind(sample2, sample3)

print(pval)
````

````python
from scipy import misc
import sys
import os
import scipy
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

os.environ['SCIPY_PIL_IMAGE_VIEWER'] = '/usr/local/bin/display'

face = misc.face() # An example image included in scipy

# Read an image from a file.
flower = misc.imread('flower.jpg')

# misc.imshow(face)
# scipy.misc.imshow is also available, but the imshow from 
# matplotlib works better.
plt.subplot(121)
plt.imshow(face)
plt.subplot(122)
plt.imshow(flower)
plt.show()

# Images are numpy arrays.

#u, s, v = np.linalg.svd(flower)

#plt.plot(s)
#plt.show()

# Specifiy the 3 axis of image data to average
# the 3 colors together.
bw_flower = np.average(flower, axis=2)
bw_face = np.average(face, axis=2)

print(flower.shape)
print(bw_flower.shape)
print(bw_face.shape)

bw_face_resize = misc.imresize(bw_face, (450, 600))

plt.subplot(131)
plt.imshow(bw_flower)
plt.subplot(132)
plt.imshow(bw_face)
plt.subplot(133)
plt.imshow(bw_face_resize)
plt.show()

# Is there anything to be learned from the SVD of these images.

bw_u, bw_s, bw_v = np.linalg.svd(bw_flower)
face_u, face_s, face_v = np.linalg.svd(bw_face_resize)

print(bw_s.shape)

s_x = np.arange(450)

plt.plot(s_x, bw_s, 'r--', s_x, face_s, 'b--')
plt.show()
````
