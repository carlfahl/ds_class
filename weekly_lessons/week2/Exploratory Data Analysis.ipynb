### Pivot tables and cross-tabulation



```python
# Using a pivot table we can view the 

# Without a values term the pivot table will show aggregation for 
# all the numeric data columns.  By default the aggregation is the mean
pd.pivot_table(train, index="Credit_History")
```

```python
# If you want to aggregate non-numeric categorical data, a aggfunc function is
# needed.
pd.pivot_table(train, values="Loan_Status", index="Credit_History", aggfunc=lambda x: x.map({'Y':1, 'N':0}).mean())
```

```python
# The index values can be non-numeric
pd.pivot_table(train, values="Loan_Status", index="Property_Area", aggfunc=lambda x: x.map({'Y':1, 'N':0}).mean())
```

```python
# More than one column may be used to sort values.
pd.pivot_table(train, values="Loan_Status", index=["Credit_History", "Property_Area"], aggfunc=lambda x: x.map({'Y':1, 'N':0}).mean())
```

A another method is the cross tabulation method

```python
pd.crosstab(train['Credit_History'], train['Loan_Status'])
```

You can use crosstab with multiple indexing values as well.

```python
# For the crosstab method the index is the first argument, values the second

pd.crosstab([train['Credit_History'], train['Gender']], train['Loan_Status'])
```
