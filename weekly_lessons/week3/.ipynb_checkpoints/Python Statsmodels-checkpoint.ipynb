#### Loading test data

statsmodels provides simple programatic access to the Rdatasets repository.  These dataset were distributed with the R programming language and are meant for teaching and development purposes.

Loading Data can be done as follows:

````python
import statsmodels.api as sm
import pandas as pd
from patsy import dmatrices

# This uses Pandas to import the data from the repo into a 
# Pandas dataframe.
df = sm.datasets.get_rdataset('sunspot.month', 'datasets').data

print(df)
````

````python
df2 = sm.datasets.get_rdataset("Guerry", "HistData").data
````

Statsmodes can also import data from other sources.  

* webuse(<data set name>, baseurl)
    * The baseurl defaults to http://www.stata-press.com/data/r11/ for release 11.  Up to release 14 is currently available.

* built in datasets
    * Listed at: http://www.statsmodels.org/stable/datasets/index.html

````python
#sm.datasets.webuse()

# List the methods of the datasets object

print(dir(sm.datasets))

# Of these that are built-in data sets: load by sm.datasets.<name>.load
# for example:

nileData = sm.datasets.nile.load_pandas()

# Here that actual Dataframe is the .data attribute.
print(nileData.data.head())
````

````python
spector_data = sm.datasets.spector.load()

print(spector_data.exog)

spector_data.exog = sm.add_constant(spector_data.exog, prepend=False)

print(spector_data.exog)

print(spector_data.endog)
````

This dataset has a lot of columns.  So we can use the Pandas method of indexing to select the columns we want.

````python
vars = ['Department', 'Lottery', 'Literacy', 'Wealth', 'Region']

df2[vars].tail()
````

````python
df2 = df2.dropna()
````

### Linear Regression

````python
y, X = dmatrices('Lottery ~ Literacy + Wealth + Region', data=df2, return_type='dataframe')
````

````python
y
````

````python
X
````

````python
mod = sm.OLS(y, X)
````

````python
res = mod.fit()
````

````python
print(res.summary())
````

````python
sm.stats.linear_rainbow(res)
````

This produced a multivaiate OLS regression of the data.  We can plot the individual 

````python
import matplotlib.pyplot as plt

print(df2[vars].tail())

sm.graphics.plot_partregress('Lottery', 'Wealth', ['Region', 'Literacy'], data=df2, obs_labels=False)
````

````python
sm.graphics.plot_partregress('Lottery', 'Region', ['Wealth', 'Literacy'], data=df2, obs_labels=False)
````

````python
sm.graphics.plot_partregress('Lottery', 'Literacy', ['Region', 'Wealth'], data=df2, obs_labels=False)
````

### Goodness of Fit

````python
resids = res.resid
sm.qqplot(resids, fit=True, line='45')
plt.show()
````
