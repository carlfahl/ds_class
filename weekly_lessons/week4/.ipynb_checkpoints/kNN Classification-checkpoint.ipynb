### kNN Classification of Iris dataset

The kNN algoithem is the simplest machine learning method.  

Nearest Neighbor methods find the datapoints with the smallest Euclidean distance 

```python
from sklearn import datasets # Easy access to iris data
import pandas as pd
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
X = np.array([[0, 0], [-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
X
```

```python
plt.plot(X[:,0], X[:,1])
plt.show()
```

We can see that the 

```python
nbrs = NearestNeighbors(n_neighbors=2, algorithm='ball_tree').fit(X)
distances, indices = nbrs.kneighbors(X)
indices
```

```python
distances
```

```python
# Calculate 3 nearest neighbors.
nbrs = NearestNeighbors(n_neighbors=3, algorithm='ball_tree').fit(X)
distances, indices = nbrs.kneighbors(X)
indices
```

In the simple example previously the training data and the test data were the same.  The closest point to each point was the point itself.

```python
iris = datasets.load_iris()

iris_df = pd.DataFrame(iris.data, columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
iris_df
```

```python
iris_df.iloc[:, [0, 1]] # Get the first two columns
```

```python
# We will do a 2 feature classification on sepal length and sepal width
X = iris_df[['sepal length', 'sepal width']]
```

```python
y = iris.target
```

```python
clf = KNeighborsClassifier(n_neighbors=15)
clf.fit(X, y)
```

```python
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
X['sepal length'].min()
```

```python
h=0.1 # The grid spacing.

x_min, x_max = X.iloc[:,[0]'sepal length'].min() - 1, X['sepal length'].max() + 1
y_min, y_max = X['sepal width'].min() - 1, X['sepal width'].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
# ravel takes the 
np.c_[xx.ravel(), yy.ravel()]
```

```python
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
Z
```

```python
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.scatter(X.iloc[:, [0]], X.iloc[:, [1]], c=y, cmap=cmap_bold)


plt.show()
```
