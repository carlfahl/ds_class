### Principle Component Analysis

Principle Component Analysis (PCA) is one of the most useful 

```python
import numpy as np
import pandas as pd
from statsmodels.multivariate.pca import PCA
import matplotlib.pyplot as plt
from sklearn import datasets # Easy access to iris data
```

We will generate normally distributed random data.

```python
x = np.random.randn(100)
print(x.shape)
x = x[:, None] # generate 100 random numbers and convert to column vector by adding a dimemsion
print(x.shape)
x
```

```python
x = x + np.random.randn(100, 100)
print(x.shape)
x
```

```python
pc = PCA(x)
```

```python
print(pc.scores)
scores_df = pd.DataFrame(pc.scores)

# Plotting the scores of all 100 loadings
scores_df.plot()
```

#### R<sup>2</sup> plot

Here we plot the R^2 value as a function of 

```python
%matplotlib inline
pc.plot_rsquare(50)
```

### PCA of the IRIS dataset

```python
iris = datasets.load_iris()

iris_df = pd.DataFrame(iris.data, columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
iris_df
```

```python
iris_pca = PCA(iris_df)
```

```python
iris_pca.scores
```

```python
iris_pca.scores.plot()
```

```python
iris_pca.plot_rsquare()
```

with real world data the R^2 converges rapidily.  We can see that 2 component fit most of the variation.  This is in great contrast to the normaly distributed random data shown before.

```python
iris_pca.loadings
```

```python
iris_factors = iris_pca.factors
iris_factors['type'] = iris.target

print(iris_factors)

# Use boolean masking from pandas to set the different colors by type of iris.

plt.plot(iris_factors[iris_factors['type'] == 0]['comp_0'], iris_factors[iris_factors['type'] == 0]['comp_1'], 'r+',
        iris_factors[iris_factors['type'] == 1]['comp_0'], iris_factors[iris_factors['type'] == 1]['comp_1'], 'b+',
        iris_factors[iris_factors['type'] == 2]['comp_0'], iris_factors[iris_factors['type'] == 2]['comp_1'], 'g+')
```

In the previous plot we can see the different iris types are fairly well disting

#### Plotting 3 PCA factors

```python
from mpl_toolkits.mplot3d import Axes3D

myFig = plt.figure()

axis = myFig.add_subplot(111, projection='3d')

axis.scatter(iris_factors[iris_factors['type'] == 0]['comp_0'], 
             iris_factors[iris_factors['type'] == 0]['comp_1'],
             iris_factors[iris_factors['type'] == 0]['comp_2'], 
             c='r', marker='o')

axis.scatter(iris_factors[iris_factors['type'] == 1]['comp_0'], 
             iris_factors[iris_factors['type'] == 1]['comp_1'],
             iris_factors[iris_factors['type'] == 1]['comp_2'], 
             c='b', marker='o')

axis.scatter(iris_factors[iris_factors['type'] == 2]['comp_0'], 
             iris_factors[iris_factors['type'] == 2]['comp_1'],
             iris_factors[iris_factors['type'] == 2]['comp_2'], 
             c='g', marker='o')
```
