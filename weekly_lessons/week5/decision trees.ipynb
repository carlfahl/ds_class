### Decision Trees



```python
from sklearn import tree, datasets
from sklearn.model_selection import train_test_split # Use to split data into training and test datasets.
from IPython.display import Image
import pydotplus
```

```python
X = [[0, 0], [1, 1], [2, 2], [3, 3], [5,5]]
Y = [0, 1, 2, 3, 4]
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X, Y)
```

#### Visualizing the tree

The tree module of scikit-learn supports exporting 

```python
dot_data = tree.export_graphviz(clf, out_file=None)
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())
```
